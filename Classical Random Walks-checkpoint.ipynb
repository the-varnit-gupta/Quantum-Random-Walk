```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1D Random Walk

```python
no_realisations=5
no_steps=10000
position=np.zeros((no_realisations,no_steps))
for i in range(no_realisations):
    for j in range(1,no_steps):
        y=np.random.rand()-0.5
        position[i][j]=position[i][j-1]+y
steps=np.arange(no_steps)

plt.figure(figsize=(10, 6))
plt.title("1D Random Walk", fontsize=15)
plt.xlabel('Steps (s)',fontsize=15)
plt.ylabel('Position (x)',fontsize=15)

for i in range(no_realisations):
    plt.plot(steps,position[i],label=f'Realization {i+1}')

plt.legend()
plt.grid(True)
plt.show()
    
```

# 2D Random Walk (10-steps)

```python
no_realisations=5
no_steps=10
positionX=np.zeros((no_realisations,no_steps))
positionY=np.zeros((no_realisations,no_steps))
for i in range(no_realisations):
    for j in range(1,no_steps):
        y=np.random.rand()-0.5
        positionX[i][j]=positionX[i][j-1]+y
        
for i in range(no_realisations):
    for j in range(1,no_steps):
        y=np.random.rand()-0.5
        positionY[i][j]=positionY[i][j-1]+y

plt.figure(figsize=(10, 6))
plt.title("2D Random Walk (10-steps)", fontsize=15)
plt.ylabel("Y-Postition",fontsize=15)
plt.xlabel("X-Postition",fontsize=15)
for i in range(no_realisations):
    plt.plot(positionX[i],positionY[i])
plt.grid(True)
plt.show()
```

# 3D Random Walk

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import the 3D plotting module

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

no_realisations=5
no_steps=3600
positionX=np.zeros((no_realisations,no_steps))
positionY=np.zeros((no_realisations,no_steps))
positionZ=np.zeros((no_realisations,no_steps))
for i in range(no_realisations):
    for j in range(1,no_steps):
        y=np.random.rand()-0.5
        positionX[i][j]=positionX[i][j-1]+y
        
for i in range(no_realisations):
    for j in range(1,no_steps):
        y=np.random.rand()-0.5
        positionY[i][j]=positionY[i][j-1]+y

for i in range(no_realisations):
    for j in range(1,no_steps):
        y=np.random.rand()-0.5
        positionZ[i][j]=positionZ[i][j-1]+y

#plt.figure(figsize=(10, 6))
plt.title("3D Random Walk", fontsize=15)
ax.set_xlabel("Y-Postition",fontsize=15)
ax.set_ylabel("X-Postition",fontsize=15)
ax.set_zlabel("Z-Postition",fontsize=15)
for i in range(no_realisations):
    plt.plot(positionX[i],positionY[i],positionZ[i])
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_walk(N, W):
    steps = np.random.uniform(-1/2, 1/2, size=(W, N))
    endpoints = np.sum(steps, axis=1)
    return endpoints

def plot_histogram(N, W, bins=50):
    endpoints = generate_random_walk(N, W)
    plt.hist(endpoints, bins=bins, density=True, alpha=0.6, label=f'N={N} Random Walk')

    # Predicted normal distribution
    mean = 0
    std_dev = np.sqrt(N/12)  # Using the central limit theorem
    x = np.linspace(-30, 30,1000)
    pdf = (1/(std_dev * np.sqrt(2*np.pi))) * np.exp(-0.5*((x - mean)/std_dev)**2)
    plt.plot(x, pdf, 'r', label='Predicted Normal Distribution')

    plt.xlabel('Endpoint')
    plt.ylabel('Probability Density')
    plt.title(f'Random Walk with {W} Walks')
    plt.legend()
    plt.show()

# Number of walks and steps
W = 10000
N_values = [1,2,3,5]


for N in N_values:
    plot_histogram(N, W)
```
